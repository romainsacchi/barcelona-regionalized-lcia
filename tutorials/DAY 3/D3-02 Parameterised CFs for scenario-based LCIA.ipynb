## Learning goals

- Build scenario-dependent characterization factors.
- Run impact results under multiple parameter sets.


## Concept

Parameterized CFs represent changing environmental conditions across scenario narratives.


```python
import pandas as pd

cf_table = pd.DataFrame([
    {'scenario': 'baseline', 'cf_co2': 1.00},
    {'scenario': 'rcp26', 'cf_co2': 0.95},
    {'scenario': 'rcp85', 'cf_co2': 1.10},
])
flow_amount = 100
cf_table['impact'] = flow_amount * cf_table['cf_co2']
cf_table
        
```
