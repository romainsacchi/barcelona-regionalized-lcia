## Concept

Criticality-weighted inventory flows provide a first-order view of supply risk hotspots.


```python
import pandas as pd

inventory = pd.DataFrame([
    {'technology': 'wind', 'material': 'copper', 'amount': 42, 'criticality_factor': 0.6},
    {'technology': 'solar', 'material': 'copper', 'amount': 28, 'criticality_factor': 0.6},
    {'technology': 'battery', 'material': 'lithium', 'amount': 15, 'criticality_factor': 0.9},
])
inventory['criticality_score'] = inventory['amount'] * inventory['criticality_factor']
inventory.sort_values('criticality_score', ascending=False)
        
```
