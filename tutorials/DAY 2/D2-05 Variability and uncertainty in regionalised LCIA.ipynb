## Learning goals

- Separate spatial variability from parameter uncertainty.
- Perform simple sensitivity checks across regions.


## Concept

Run the same model with multiple plausible regions to understand spread and robustness.


```python
import pandas as pd

scenario_scores = pd.DataFrame([
    {'region_case': 'Ecoregion A', 'score': 14.2},
    {'region_case': 'Ecoregion B', 'score': 17.9},
    {'region_case': 'Ecoregion C', 'score': 12.8},
])
print('Min:', scenario_scores['score'].min())
print('Max:', scenario_scores['score'].max())
print('Range:', scenario_scores['score'].max() - scenario_scores['score'].min())
scenario_scores
        
```
