## Learning goals

- Interpret category-specific spatial sensitivity.


## Concept

Different categories can respond differently to spatial detail (water, biodiversity, human health).


```python
import pandas as pd

results = pd.DataFrame([
    {'category': 'freshwater use', 'score_global': 10.0, 'score_regionalized': 18.0},
    {'category': 'biodiversity', 'score_global': 2.5, 'score_regionalized': 4.1},
    {'category': 'human health', 'score_global': 5.2, 'score_regionalized': 5.8},
])
results['ratio_regionalized_to_global'] = results['score_regionalized'] / results['score_global']
results
        
```

Identify which category is most sensitive to regionalization and justify with the ratio.

```python
most_sensitive = results.sort_values('ratio_regionalized_to_global', ascending=False).iloc[0]
print('Most sensitive category:', most_sensitive['category'])
print('Ratio:', most_sensitive['ratio_regionalized_to_global'])
        
```
