## 4) Import checks in the current kernel

The goal is to verify whether the notebook kernel can see the main course packages.

Note:
- If `bw2data` prints a warning about missing sparse solvers, read it carefully.
- If one or more imports fail now, that is acceptable before the `bw2` environment is created.


````python
import importlib

modules = ['bw2data', 'bw2calc', 'bw2io', 'edges']

for name in modules:
    try:
        module = importlib.import_module(name)
        version = getattr(module, '__version__', 'unknown')
        print(f'{name}: OK | version={version}')
    except Exception as exc:
        print(f'{name}: FAIL | {type(exc).__name__}: {exc}')
````

## 5) Check the bundled BAFU workbook

The Day 1 import notebook uses the workbook already shipped with this repository:

- `data/lci-bafu.xlsx`


````python
from datetime import datetime
from pathlib import Path

bafu_path = Path('data/lci-bafu.xlsx')
print('Expected path:', bafu_path.resolve())
print('Found:', bafu_path.exists())
if bafu_path.exists():
    size_mb = bafu_path.stat().st_size / 1024 / 1024
    mtime = datetime.fromtimestamp(bafu_path.stat().st_mtime)
    print('Size [MB]:', round(size_mb, 2))
    print('Last modified:', mtime.isoformat(timespec='seconds'))
    if size_mb < 10:
        print('Warning: the workbook is smaller than expected; verify that it is complete.')
````

## Troubleshooting quick list

- `conda: command not found`: reopen the shell, and on Windows make sure you are in `Miniforge Prompt`.
- Wrong kernel: switch the notebook kernel to the environment where `bw2` was installed.
- `edges` import fails: the `bw2` environment was not created correctly, or the kernel is not using it.
- Sparse solver warning during `bw2data` import: follow the environment file and reinstall the missing solver dependencies.
- `data/lci-bafu.xlsx` is missing: update the repository before Day 1.
